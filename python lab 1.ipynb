```python


import numpy as np
import matplotlib.pyplot as plt
import labx as mx

student_id = 2600562


x = np.arange(-10, 10.001, 0.001)
y = mx.lab_1_problem_1(x, student_id)


plt.figure(figsize=(12, 6))
plt.plot(x, y, 'b-', linewidth=1.5)
plt.axhline(y=1, color='r', linestyle='--', linewidth=2, label='y = 1')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Overall view - Finding intersections with y = 1')
plt.grid(True, alpha=0.3)
plt.xlim(-10, 10)
plt.legend()
plt.show()




plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', linewidth=2)
plt.axhline(y=1, color='r', linestyle='--', linewidth=2, label='y = 1')
plt.xlim(-7.5, -6.5)
plt.ylim(0.5, 1.5)
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.title('Intersection Point 1 - Zoomed View (Left side)', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()


plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', linewidth=2)
plt.axhline(y=1, color='r', linestyle='--', linewidth=2, label='y = 1')
plt.xlim(-7.2, -6.8)  # Even more zoomed in
plt.ylim(0.8, 1.2)    # Even more zoomed in around y=1
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.title('Intersection Point 1 - Further Zoomed (Left side)', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', linewidth=2)
plt.axhline(y=1, color='r', linestyle='--', linewidth=2, label='y = 1')
plt.xlim(1.5, 2.5)
plt.ylim(0.5, 1.5)
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.title('Intersection Point 2 - Zoomed View (Right side)', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()


plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', linewidth=2)
plt.axhline(y=1, color='r', linestyle='--', linewidth=2, label='y = 1')
plt.xlim(1.8, 2.2)
plt.ylim(0.9, 1.1)
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.title('Intersection Point 2 - Further Zoomed (Right side)', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()


print("Based on the zoomed graphs, you can read the intersection points:")
print("First intersection point (left side): x ≈ ___ .__")
print("Second intersection point (right side): x ≈ ___ .__")
```
